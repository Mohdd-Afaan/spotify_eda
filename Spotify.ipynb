10 most popular songs

```python
most_popular = df_tracks.query('popularity>90', inplace=False).sort_values('popularity', ascending= False)
most_popular[:10]
```

Setting index as "release_date" column and changing the values to date time format

```python
df_tracks.set_index("release_date", inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index, format="mixed")
df_tracks.head(5)
```

coverting the "duration_ms" from milliseconds into seconds

```python
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace= True , axis=1)
df_tracks.head(5)
```

Vizualization of correlation (pearson)

```python
corr_df = df_tracks.drop(["key", "mode", "explicit", "id", "name", "artists", "id_artists"], axis=1).corr(method="pearson")
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="black")
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
# creating new column dates 
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
year = df_tracks.dates.dt.year
```

vizulaizations

Distribution plot

```python
sns.displot(year, discrete=True, aspect=2, height=5, kind='hist').set(title= "Number of songs per year")
```

Bar plot

```python
total_dr= df_tracks.duration
fig, ax = plt.subplots(figsize= (18,7))
fig = sns.barplot(x=year, y=total_dr, ax=ax, errwidth=False).set(title="year vs duration")
plt.xticks(rotation=90)
```

line plot

```python
total_dr= df_tracks.duration
sns.set_style(style="whitegrid")
fig, ax = plt.subplots(figsize= (18,7))
fig = sns.lineplot(x=year, y=total_dr, ax=ax).set(title="year vs duration")
plt.xticks(rotation=60)
```
